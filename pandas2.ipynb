```python
df.columns,len(df.columns)
```

```python
df.columns=[x.lower() for x in df.columns]
print(df.columns)
```

```python
type(df['age'])
```

```python
df['family_size']=df['sibsp']+df['parch']
print(df.columns)
```

```python
df['fare_per_person']=df['fare']/(df['family_size']+1)
print(df.columns)
print(df['fare'],df['family_size'],df['fare_per_person'])
```

```python
df['is_alone']=df['family_size'].apply(lambda x:1 if x==0 else 0)
print(df['is_alone'])
```

```python
df['is_alone_bool']=(df['family_size'].astype('bool')).astype('int')
print(df['is_alone_bool'])
```

```python
df['is_alone_condition']=(df['family_size']==0).astype('int')
print(df[['family_size','is_alone_condition']])
```

```python
df=df.rename(columns={'sibsp':'sibling/spouse abroad'})
print(df.columns)
```

```python
df=df.rename(columns={'pclass':'passenger_class','sex':'gender'})
print(df.columns)
```

```python
df['is_adult']=(df['age']>=18).astype('int')
print(df[['age','is_adult']])
```
